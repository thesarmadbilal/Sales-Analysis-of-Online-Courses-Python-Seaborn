```python
import numpy as py 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Read file and encoding to avoid error
df = pd.read_csv('Online_Tech_Courses_Data2023.csv', encoding = 'unicode_escape')
```

```python
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
## On the basis of Gender= "Male, Female"
vx = sns.countplot(x = 'Gender', data = df)

for bars in vx.containers:
    vx.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
```

```python
sns.countplot(data = df, x ='Age Group', hue = 'Course_Name')
filtered_df = df[~df['Age Group'].isin(['0-17', '55+'])]

sns.set(rc={'figure.figsize':(15,5)})

# Show the plot
plt.show()
```

```python
# Specify the courses to include in the plot
selected_courses = ['Data Science', 'Python Programming']

# Filter out the age groups '0-17' and '55+' and select specific courses
filtered_df = df[(~df['Age Group'].isin(['0-17', '55+'])) & (df['Course_Name'].isin(selected_courses))]

# Create the countplot with the filtered DataFrame
ax = sns.countplot(data=filtered_df, x='Age Group', hue='Course_Name')

sns.set(rc={'figure.figsize': (15, 5)})

# Show the plot
plt.show()
```

```python
##No of Orders by Region
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(20)
sns.set(rc ={'figure.figsize': (15,5)})


sns.barplot(data = sales_state, x = 'State', y = 'Orders', palette = 'ch:s=.25,rot=-.25')
```

```python
## count of people by Occupation 
ox = sns.countplot(data = df, x = 'Occupation')

for bars in ox.containers:
    ox.bar_label(bars)
```

```python
#Course on the basis of Amount
sales_state = df.groupby(['Course_Name'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(20)
sns.set(rc ={'figure.figsize': (15,5)})


sns.barplot(data = sales_state, x = 'Course_Name', y = 'Amount', palette = 'light:b')
```

```python
#Comparison of courses by Occupation
sns.countplot(data = df, x ='Occupation',  hue = 'Course_Name')
```

```python
#Orders on the basis of courses
sales_state = df.groupby(['Course_Name'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(20)
sns.set(rc ={'figure.figsize': (15,5)})


sns.barplot(data = sales_state, x = 'Course_Name', y = 'Orders', palette = 'light:b')
```
